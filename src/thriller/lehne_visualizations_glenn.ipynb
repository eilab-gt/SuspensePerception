```python
def calculate_deltas(data):
    return [data[i+1] - data[i] for i in range(len(data)-1)]

delta_llama = calculate_deltas(suspense_llama3_8B)
delta_human = calculate_deltas(suspense_humans)
```

```python
def get_direction(deltas):
    return [1 if d > 0 else (-1 if d < 0 else 0) for d in deltas]

direction_llama = get_direction(delta_llama)
direction_human = get_direction(delta_human)
```

```python
agreement = [1 if h == l else 0 for h, l in zip(direction_human, direction_llama)]
```

```python
print("Delta Llama:", delta_llama)
print("Delta Human:", delta_human)
print("Direction Llama:", direction_llama)
print("Direction Human:", direction_human)
print("Agreement:", agreement)
```

```python
import matplotlib.pyplot as plt

def classify_value(x):
    if x > 5.5:
        return 1
    elif x < 5.5:
        return -1
    else:
        return 0

# Apply the function to both datasets
classified_llama = [classify_value(x) for x in suspense_llama3_8B]
classified_human = [classify_value(x) for x in suspense_humans]
```

```python
agreement = []
for (l, h) in zip(classified_llama, classified_human):
  if l == h:
    print("yes")
    agreement.append(1)
  else:
    print("no")
    agreement.append(0)
print(agreement)
```
