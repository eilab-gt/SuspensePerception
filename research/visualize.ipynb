```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
filepath = "../outputs/lehne_experiment/meta-llama_Meta-Llama-3.1-70B-Instruct-Turbo/results.parquet"
```

```python
df = pd.read_parquet(filepath, engine="pyarrow")
response = df.loc[:, "response"][0]
suspense_ratings = [int(value) for value in response.values()]
```

```python
# Generate x-axis labels
text_segments = list(range(1, len(suspense_ratings) + 1))

# Create the plot
plt.figure(figsize=(14, 7))

# Plot human ratings
# plt.plot(text_segments, suspense_ratings_human, label='Human Ratings', color='blue', marker='o')

# Plot llama ratings
plt.plot(text_segments, suspense_ratings, label='LLM Ratings', color='blue', marker='x')

# Labeling the plot
plt.xlabel('Text Segment')
plt.ylabel('Suspense Rating')
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```
